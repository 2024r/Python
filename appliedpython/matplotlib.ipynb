```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,6,7,8,9,5]
y=[2,4,6,16,8,10,12,14,18]
plt.plot(x, y)
```

```python
plt.plot(x,y)
plt.title("nummber count frequency ")
plt.xlabel('frequency')
plt.ylabel("number count")
```

```python
plt.plot(x,y, color='red', linestyle='dotted')
plt.title("nummber count frequency ")
plt.xlabel('frequency')
plt.ylabel("number count")
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y=['java', 'python', 'c++', 'javascript', 'ruby']
```

```python
plt.pie(x, labels=y)
plt.show()
```

```python
ex=[1, 2, 3, 4, 5]
plt.pie(ex, labels=y ,explode=(0.1, 0, 0, 0, 0), autopct='%1.1f%%', shadow=True)
plt.show()
```

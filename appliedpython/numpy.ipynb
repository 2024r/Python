slicing in numpy array

```python
import numpy as np  # Import numpy with an alias
arr1 = np.arange(10)  # Create an array with values from 0 to 9
arr2 = np.arange(12).reshape(3, 4)  # Corrected 'rearange' to 'reshape'
print(arr1)  # Output the first array
print(arr2)  # Output the second array with reshaped dimensions
```

```python
arr1
```

```python
arr2
```

```python
# To slice numpy arrays, use arr2[row_slice, col_slice] syntax
# Example: select rows in reverse order and columns in steps of -3
result = arr2[-1::-2, -1::-3]
print(result)
```

```python
result= arr2[1::, ::2]  # Slicing the array
print(result)  # Output the sliced result
```

```python
arr2
```

```python
print(arr2[0, :])
print(arr2[ :0]) 
print(arr2[1:,1:])
print(arr2[::-2,:2:]) # Output the second row of arr2
print()  # Output the original array for reference
```

3d array slicing

```python
arr3=np.arange(24).reshape( 2,4,3 )  # Create a new array with values from 0 to 11 reshaped to 3 rows and 4 columns
arr3
```

```python
print(arr3[0], end='\n\n')  # Output the first block of arr3

print(arr3[0, 0], end='\n\n')  # Output the first row of the first block

print(arr3[1, :3], end='\n\n')  # Output the first three rows of the second block

print(arr3[0, :2, ::2], end='\n\n')  # Output the first two rows and every second column of the first block
```
